```python
import pandas as pd

def clean_twitter_data(src, dst):
    values = []
    with open(src) as f:
        
        for i, line in enumerate(f):
            if i == 0:
                colnames = line
            else:
                values.append(line.split(',')[:4])
    data = pd.DataFrame(values, columns = colnames.split(','))
    return data

clean_twitter_data('/mnt/ml-team/minerva/unblackboxing_webinar/Sentiment Analysis Dataset.csv',
                  '/mnt/ml-team/minerva/unblackboxing_webinar/twitter_sentiment.csv')
```
